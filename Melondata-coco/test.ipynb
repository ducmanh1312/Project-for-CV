```python
import json
import cv2
import re
import os

# Load the JSON file
paths = ('test', 'valid', 'train')
labels = []
for path in paths:
    print(path +'/_annotations.coco.json')
    with open(path +'/_annotations.coco.json', 'r', encoding='utf-8') as file:
        data = json.load(file)

    # Print the general info section
    info = data.get("info", {})
    print("General Info:")
    print(f"Year: {info.get('year')}")
    print(f"Version: {info.get('version')}")
    print(f"Description: {info.get('description')}")
    print(f"Date Created: {info.get('date_created')}")
    print()

    # Print the categories
    print("Categories:")
    categories = data.get("categories", [])
    for category in categories:
        print(f"ID: {category['id']}, Name: {category['name']}, Supercategory: {category['supercategory']}")
    print()

    # Print image information
    print("Images:")
    images = data.get("images", [])
    print(images[0]['file_name'])
    print(images)
    for image in images:
        print(f"ID: {image['id']}, File Name: {image['file_name']}, Width: {image['width']}, Height: {image['height']}, Date Captured: {image['date_captured']}")
        print("-")
    print()

    # Print the anotation
    print("Anotation:")
    anotations = data.get("annotations", [])
    
    for anotation in anotations: 
        print(f"ID: {anotation['id']},image_id: {anotation['image_id']}, category_id: {anotation['category_id']}, bbox:{anotation['bbox']}, area:{anotation['area']}, segmentation:{anotation['segmentation']}, iscrowd:{anotation['iscrowd']}")
        image_id = anotation['image_id']
        image = cv2.imread(path + '/' + str(images[image_id]['file_name']))

        # Extract numbers using map
        x_start, y_start, width, height = map(int, anotation['bbox'])
        print()

        # # Assign to variables
        x_end = x_start + width
        y_end = y_start + height

        # Cắt ảnh
        cropped_image = image[y_start:y_end, x_start:x_end]

        # Hiển thị ảnh đã cắt
        # cv2.imshow('Cropped Image', cropped_image)
        # cv2.waitKey(0)

        # Đóng cửa sổ hiển thị
        cv2.destroyAllWindows()

        # Lưu ảnh đã cắt (nếu muốn)
        category_id = anotation['category_id']
        folder_path = "data/" + path
        file_name = str(anotation['id']) + '_cropped_image.jpg'

        # print(categories[category_id]['name'])
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)
    
        print(folder_path + '/' + str(anotation['id']) + 'cropped_image.jpg' )
        cv2.imwrite( folder_path + '/' + file_name, cropped_image)

        # Lưu nội dung label
        output_file = "data/image_metadata.json"

        label = {
            "type": path,
            "filename": file_name,
            "category": categories[category_id],
        }
        labels.append(label)
        
        with open(output_file, 'w', encoding='utf-8') as file:
            json.dump(labels, file, ensure_ascii=False, indent=4)

        print(f"Metadata đã được lưu vào file {output_file}")

```
